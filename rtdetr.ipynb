```python
import os

base_path = "/kaggle/input/ua-detrac-dataset/content/UA-DETRAC/DETRAC_Upload"

train_image_path = os.path.join(base_path, 'images/train')
val_image_path = os.path.join(base_path, 'images/val')

train_label_path = os.path.join(base_path, 'labels/train')
val_label_path = os.path.join(base_path, 'labels/val')
print(f"Train images path: {train_image_path}")
print(f"Validation images path: {val_image_path}")
print(f"Train labels path: {train_label_path}")
print(f"Validation labels path: {val_label_path}")

import yaml

data = {
    'train': ['/kaggle/input/ua-detrac-dataset/content/UA-DETRAC/DETRAC_Upload/images/train', '/kaggle/input/ua-detrac-dataset/content/UA-DETRAC/DETRAC_Upload/labels/train'],
    'val': ['/kaggle/input/ua-detrac-dataset/content/UA-DETRAC/DETRAC_Upload/images/val', '/kaggle/input/ua-detrac-dataset/content/UA-DETRAC/DETRAC_Upload/labels/val'],
    'nc': 9
}

# Specify the file path for saving the YAML file
file_path = 'data.yaml'

# Write data to YAML file
with open(file_path, 'w') as file:
    yaml.dump(data, file)

print(f"YAML file saved as {file_path}")
```
